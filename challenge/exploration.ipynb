## 2. Features Generation

### 2.a. Period of Day

```python
from datetime import datetime

def get_period_day(date):
    date_time = datetime.strptime(date, '%Y-%m-%d %H:%M:%S').time()
    morning_min = datetime.strptime("05:00", '%H:%M').time()
    morning_max = datetime.strptime("11:59", '%H:%M').time()
    afternoon_min = datetime.strptime("12:00", '%H:%M').time()
    afternoon_max = datetime.strptime("18:59", '%H:%M').time()
    evening_min = datetime.strptime("19:00", '%H:%M').time()
    evening_max = datetime.strptime("23:59", '%H:%M').time()
    night_min = datetime.strptime("00:00", '%H:%M').time()
    night_max = datetime.strptime("4:59", '%H:%M').time()
    
    if(date_time > morning_min and date_time < morning_max):
        return 'mañana'
    elif(date_time > afternoon_min and date_time < afternoon_max):
        return 'tarde'
    elif(
        (date_time > evening_min and date_time < evening_max) or
        (date_time > night_min and date_time < night_max)
    ):
        return 'noche'
```

```python
data['period_day'] = data['Fecha-I'].apply(get_period_day)
```

### 2.b. High Season

```python
from datetime import datetime

def is_high_season(fecha):
    fecha_año = int(fecha.split('-')[0])
    fecha = datetime.strptime(fecha, '%Y-%m-%d %H:%M:%S')
    range1_min = datetime.strptime('15-Dec', '%d-%b').replace(year = fecha_año)
    range1_max = datetime.strptime('31-Dec', '%d-%b').replace(year = fecha_año)
    range2_min = datetime.strptime('1-Jan', '%d-%b').replace(year = fecha_año)
    range2_max = datetime.strptime('3-Mar', '%d-%b').replace(year = fecha_año)
    range3_min = datetime.strptime('15-Jul', '%d-%b').replace(year = fecha_año)
    range3_max = datetime.strptime('31-Jul', '%d-%b').replace(year = fecha_año)
    range4_min = datetime.strptime('11-Sep', '%d-%b').replace(year = fecha_año)
    range4_max = datetime.strptime('30-Sep', '%d-%b').replace(year = fecha_año)
    
    if ((fecha >= range1_min and fecha <= range1_max) or 
        (fecha >= range2_min and fecha <= range2_max) or 
        (fecha >= range3_min and fecha <= range3_max) or
        (fecha >= range4_min and fecha <= range4_max)):
        return 1
    else:
        return 0
```

```python
data['high_season'] = data['Fecha-I'].apply(is_high_season)
```

### 2.c. Difference in Minutes

```python
def get_min_diff(data):
    fecha_o = datetime.strptime(data['Fecha-O'], '%Y-%m-%d %H:%M:%S')
    fecha_i = datetime.strptime(data['Fecha-I'], '%Y-%m-%d %H:%M:%S')
    min_diff = ((fecha_o - fecha_i).total_seconds())/60
    return min_diff
```

```python
data['min_diff'] = data.apply(get_min_diff, axis = 1)
```

### 2.d. Delay

```python
threshold_in_minutes = 15
data['delay'] = np.where(data['min_diff'] > threshold_in_minutes, 1, 0)
```

## 3. Data Analysis: Second Sight

### How is the delay rate across columns?

```python
def get_rate_from_column(data, column):
    delays = {}
    for _, row in data.iterrows():
        if row['delay'] == 1:
            if row[column] not in delays:
                delays[row[column]] = 1
            else:
                delays[row[column]] += 1
    total = data[column].value_counts().to_dict()
    
    rates = {}
    for name, total in total.items():
        if name in delays:
            rates[name] = round(total / delays[name], 2)
        else:
            rates[name] = 0
            
    return pd.DataFrame.from_dict(data = rates, orient = 'index', columns = ['Tasa (%)'])
```

```python
destination_rate = get_rate_from_column(data, 'SIGLADES')
destination_rate_values = data['SIGLADES'].value_counts().index
plt.figure(figsize = (20,5))
sns.set(style="darkgrid")
sns.barplot(destination_rate_values, destination_rate['Tasa (%)'], alpha = 0.75)
plt.title('Delay Rate by Destination')
plt.ylabel('Delay Rate [%]', fontsize=12)
plt.xlabel('Destination', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

# Model Selection Analysis

In this case, the following models were evaluated:

*   **Logistic Regression** (all features)
*   **XGBoost** (all features)
*   **Logistic Regression** (top 10 features, unbalanced)
*   **XGBoost** (top 10 features, unbalanced)
*   **Logistic Regression** (top 10 features, balanced data)
*   **XGBoost** (top 10 features, balanced data)

### Initial Assessment
At first glance, the metrics indicate that the **Logistic Regression** model performs better in terms of precision and recall. Additionally, this model trains faster, making it the best option for maintenance and future retraining.

However, since the training dataset is extensive, the preferred option is to apply a feature engineering strategy (selecting the top 10 features), as applied in the latter four models.

### The Imbalance Problem
The data is clearly unbalanced. Consequently, it is likely that the model is not learning to strictly discern between the two classes, but rather becoming "lazy" when deciding between one class or another.

This behavior is evident in the last four models. When applying a strategy to compensate for the imbalance, some metrics (such as precision) decrease, while others increase. This suggests that the model is actually learning to differentiate.

For example, looking at the **XGBoost** case:

#### Balanced Data
**Class 0**
*   Precision: 0.88
*   Recall: 0.52
*   F1-Score: 0.66

**Class 1 (Delays)**
*   Precision: 0.25
*   **Recall: 0.69**
*   F1-Score: 0.37

#### Unbalanced Data
**Class 0**
*   Precision: 0.81
*   Recall: 1.00
*   F1-Score: 0.90

**Class 1 (Delays)**
*   Precision: 0.76
*   **Recall: 0.01**
*   F1-Score: 0.01

In the unbalanced case, although the model appears to detect the majority class "better," the recall disparity is extreme. This indicates the model is not learning patterns but is simply taking the easiest statistical path. In the balanced data case, the recall value for the delayed class increases significantly.

### Business Impact
The ability to correctly differentiate between classes allows us to adopt better strategies for delayed flights. For example:
*   Offering a discount on the customer's next trip.
*   Providing special snacks or amenities during the flight.
*   Better management of ground staff resources to assist passengers.

These actions help maintain customer loyalty and mitigate the negative experience of a delay.

### Decision: Logistic Regression vs. XGBoost
Since both models yield similar metrics after balancing, we can opt for the model with **lower complexity and easier maintenance**, which is **Logistic Regression**.

### Summary
The **Logistic Regression model trained with the top 10 features using balanced data** is selected, as it demonstrates the best ability to correctly discern between the two classes.
